```python
from param_graph.seq_to_net import seq_to_net
from param_graph.visualize import draw_graph
import torch.nn as nn
from pprint import pprint
import torch
import networkx as nx
from param_graph.graph_types import BIAS_NODE_TYPES, EdgeType
```

```python
model = nn.Sequential(
    nn.Linear(5, 10),
    nn.ReLU(),
    nn.Linear(10, 10),
    nn.ReLU(),
    nn.Linear(10, 5),
    nn.ReLU(),
    nn.Linear(5, 1),
)
global_graph = seq_to_net(model)
sequential_title = ',\n'.join([str(type(module)).split('.')[-1].strip(">'") for module in model])
some_path = './graph-app/backend/static/graph1.html'
draw_graph(global_graph, dim='3d',title=sequential_title, display_inline=True)#, save_path=some_path)
#print(global_graph.get_bipartite_adjacency_matrix())
```

```python
model = nn.Sequential(
    nn.Conv2d(3,4, 3),
    nn.ReLU(),
    nn.Conv2d(4, 4, 3),
    nn.ReLU(),
    nn.Flatten(),
    nn.Linear(4*4, 10),
    nn.ReLU(),
    nn.Linear(10, 1),

)
print('total params', sum(p.numel() for p in model.parameters()))
global_graph = seq_to_net(model)

print("total edges", len(global_graph.edges))
# save the graph
sequential_title = ',\n'.join([str(type(module)).split('.')[-1].strip(">'") for module in model])
some_path = './graph-app/backend/static/graph2.html'
# draw_graph(global_graph, dim='3d',title=sequential_title, display_inline=True)
draw_graph(global_graph, dim='3d',title=sequential_title, display_inline=False, save_path=some_path)

# [edge['edge_obj'] for edge in global_graph.edges.values() if edge['edge_obj'].features.edge_type == EdgeType.CONV_BIAS]
```
